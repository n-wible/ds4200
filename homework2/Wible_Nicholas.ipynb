    Generate a data visualization to show the relationship between max temperature and weather type.

```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

```python
# Display temperature extremes x weather type using box plot
boxplot_weather = alt.Chart(weather).mark_boxplot().encode(
    x = 'weather:N',
    y = 'temp_max:Q',
    color = 'weather:N'
).properties(title = 'Temperature Distribution Across Weather Conditions')
boxplot_weather
```

##### Question 2 (10 points)

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
# Compare max & min temperatures x weather condition w scatter plot
scatter_temp = alt.Chart(weather).mark_point(filled=True).encode(
    x = 'temp_min:Q',
    y='temp_max:Q',
    color = 'weather:N',
    tooltip = ['date:T', 'weather:N', 'temp_min:Q', 'temp_max:Q']
).properties(title = "Min vs Max Temps by Weather")
scatter_temp
```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels.

```python
# Faceted scatter plot - min vs max temperature, color by wind speed
weather_hover = alt.selection_point(fields=["date"], on="mouseover", nearest=True, empty=False)

scatter = alt.Chart(weather).mark_point(filled=True).encode(
    x = 'temp_min:Q',
    y = 'temp_max:Q',
    color = alt.Color('wind:Q', scale=alt.Scale(scheme='accent')),
    opacity = alt.condition(weather_hover, alt.value(1.5), alt.value(1.5)),
    tooltip = ['temp_min:Q', 'temp_max:Q', 'wind:Q']).add_params(weather_hover)

faceted_chart = scatter.facet(
    facet = 'weather:N',
    columns = 3).properties(title = 'Min vs Max Temperature by Weather Type (Colored by Wind Speed)').interactive()
faceted_chart
```

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Note, just use the `date` as X attribute, no need to aggragate for years and you will get a wide figure. Include necessary title and labels.

```python
# display avg temp over time, mark types
# daily average temperature
weather_avg = weather.copy()
weather_avg["temp_avg"] = (weather_avg["temp_min"] + weather_avg["temp_max"]) / 2
base = alt.Chart(weather_avg).encode(
    x = 'date:T'
)

# daily line
daily = base.mark_line(opacity=0.4).encode(
    y = alt.Y('temp_avg:Q', title='Avg Temperature (°C)')
)

# mean line (rolling)
rolling = base.transform_window(
    rolling_avg = 'mean(temp_avg)',
    frame = [-14, 14]
).mark_line(size=2).encode(
    y = 'rolling_avg:Q'
)

# mo avg points
monthly = base.transform_timeunit(month='yearmonth(date)').transform_aggregate(
    mean_temp = 'mean(temp_avg)',
    groupby = ['month']
).mark_circle(opacity=0.7).encode(
    x = 'month:T',
    y = 'mean_temp:Q',
    tooltip = ['month:T', 'mean_temp:Q']
)

# merge layers
combined_chart = (daily + rolling + monthly).properties(
    title = 'Daily Avg Temperature, with Rolling Mean & Monthly Summary')
combined_chart
```
