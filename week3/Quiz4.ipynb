## Task abstraction on page 9 of More_abstraction_and_visual_encoding.pdf

Our campus bike-share program needs a 1-page overview for the dean.
We suspect weekend demand is rising near dorms; marketing wants to
compare members vs. casual riders, and facilities wants to know which
stations hit capacity. Include trends over the last 12 months, weather
effects (rain/temp), and show which 5 stations drive most growth. The
dean wants a clean version she can show to alumni.”
In the quiz, submit a list to descrive no more than 3 visulizations to
address all their requirement.

1) Scatter Chart showing the weather and temperature effects by rider type. It would show the rides per day (y) and the average temp (x) with the weather being rain/no rain (color-coded scatter points) show the relationships of Weather and Temperature on ridership.
2) Multiple Line Chart showing the Top 5 Growth Stations over the last 12 months. This would show the number of rides (y) across 12 months (x) with the top 5 stations (color-coded lines) showing the growth.
3) A Multiple Line Chart showing the number of rides (y) across multiple months (x) with the rider type of member/casual (color-coded lines) showing the growth of ridership by rider type.

## One example figure from online to fulfill a high level task

There are three levels of the action:
- High-level: Analyze, either to consume existing data or to also
produce additional data or identity the overall trend in the data
- Mid-level: Search or filter, looking for target or location
low-level: Query, like compare or summarize, or look for a specific
pattern
- The choices at each of these three levels are independent from each other,
and it’s usually useful to describe actions at all three of them

High Level example: Exports/Imports line chart to trade balance (exports-imports)
