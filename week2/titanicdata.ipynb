```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# dataframe
df = pd.read_csv('titanic.csv')

# viz 1 (bar plot) - Survival Rate x Passenger Class
survival_by_class = df.groupby('Pclass')['Survived'].mean()
sns.barplot(x=survival_by_class.index, y=survival_by_class.values)
plt.title('Survival Rate by Passenger Class', fontsize=24)
plt.ylabel('Survival Rate', fontsize = 18)
plt.xlabel('Class of Passenger', fontsize = 18)
```

## Survival Rate by Passenger Class

The Titanic dataset contains a total of 891 passengers, covering info about their survival, demographics and ticket info. 

This bar chart shows a large class-based disparity in the rate of survival in the tragedy, as 1st-class passengers had a survival rate of 63%, while only 24% of 3rd-class passengers survived. 

The visualization shows how socioeconomic status (by class on ticket) influenced the chance of survival, likely from preferential treatment and location of theirr rooms. 

Reflects an un-spoken 'wealthy first' protocol that was implemented on the ship during the event. 

```python
# viz 2 (box plot) - age x survival
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Age Distribution by Passenger Survival', fontsize=24)
plt.ylabel('Age (yrs)', fontsize = 18)
plt.xlabel('Survived (1=Yes, 0=No)', fontsize = 18)
```

## Age Distribution by Survival Status

The box plot compares the age distributions of the survivors and those who died. 

The median age is similar for both, an average of around 30 years, but survivors were slightly younger on average. 

The plot shows that children had very strong survival chances, as the lower quartile is much younger.

Some elderly passengers (70-80yrs) did survive and die, but there isnt a big enough survey side to determine if there was any preferential treatment. 

The plot ranges are similar, and age alone wasn't a primary factor for determining survival, but instead was a much smaller role in whether they got on lifeboats.

```python
# viz 3 (count plot) - passenger count x sex x survival
sns.countplot(data=df, x='Sex', hue='Survived', palette='husl')
plt.title('Passenger Survival by Sex', fontsize=24)
plt.ylabel('Passenger Count', fontsize = 18)
plt.xlabel('Sex', fontsize = 18)
plt.legend(title='Survived', labels=['No', 'Yes'])
```

## Survival Rate by Sex of Passengers

The plot shows very different survival rates between male and female passengers on the titanic in 1912. 

The dataset had 577 males and 314 females represented, and despite this, males died much more on average.

The plot shows that 233 of 314 females survived at a rate of 74.2%. 

Out of the 577 males on the ship, only 109 survived, representing a 18.89% survival rate. 

This contrast echoes the famous 'women and children first' protocol which was implemented that night. 

The data shows that sex was the primary and strongest determination of survival in the Titanic disaster.
